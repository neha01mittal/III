```python
users_by_country = pd.read_csv('data/num_users_by_country.csv', encoding= 'unicode_escape')
```

```python
users_by_country.head()
```

```python
# Load users per country

users_by_country['Country or area'] = users_by_country['Country or area'].str.strip()
```

```python
users_by_country.head()
```

```python
users_by_country.count()
```

```python
sum_court = sum_court.reset_index()
```

```python
# Left join google request data with total users per country dataframe
google_requests_users = pd.merge(sum_court,
        users_by_country,
        left_on="Country",
        right_on="Country or area",
        how="left")
```

```python
google_requests_users[:10]
```

```python
google_requests_users = google_requests_users.dropna()
```

```python
google_requests_users.count()
```

```python
# Keep the useful columns only. Filter the rest
google_requests = google_requests_users[['Country', 'Items Requested To Be Removed', 'Internet users', 'Population']]
```

```python
google_requests['Internet users'] = google_requests['Internet users'].str.replace(',', '')
google_requests['Population'] = google_requests['Population'].str.replace(',', '')
```

```python
google_requests['Internet users'] = pd.to_numeric(google_requests['Internet users'], errors='coerce')
google_requests['Population'] = pd.to_numeric(google_requests['Population'], errors='coerce')
```

```python
type(google_requests['Internet users'][0])
```

```python
# Calculate transparency metric:
google_requests['transparency_metric'] = 1000000 * google_requests['Items Requested To Be Removed']/google_requests['Internet users']
```

```python
google_requests_sorted = google_requests.sort_values(by=['transparency_metric'], ascending=False)
```
