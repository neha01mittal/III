```python
info_reqs_msft = pd.read_csv('../data/Microsoft_Removal_Requests.csv', encoding = "utf-8")
info_reqs_msft.head()
info_reqs_msft = info_reqs_msft.drop(['Country'], axis=1)
```

```python
sum_court_year.reset_index(inplace=True)
sum_court_year_pivoted = sum_court_year.pivot(index='Country.1', 
                                              columns = 'Year', values= 'Removal_Requests').reset_index()

```

```python
sum_court_year_pivoted.isna().sum()
```

```python
sum_court_year_pivoted_reduced = sum_court_year_pivoted.dropna()
```

```python
sum_court_year_pivoted_reduced.head()
```

```python
users_by_country_year = users_by_country_year.drop(['1996_y', '1997_y', '1998_y', '1999_y', '2000_y',
                                                    '2001_y', '2002_y', '2003_y', '2004_y', '2005_y', '2006_y', '2007_y', '2008_y', 
                                                    '2009_y', '2010_y', '2011_y',
                                                   '2012_y', '2013_y', '2014_y'], axis=1)


```

```python
users_by_country_year.head()
```

```python
msft_requests_users = pd.merge(sum_court_year_pivoted_reduced,
        users_by_country_year,
        left_on="Country.1",
        right_on="Country Name",
        how="inner")
```

```python
len(sum_court_year_pivoted_reduced)
len(users_by_country_year)
```

```python
sum_court_year_pivoted_reduced.columns
```

```python
msft_requests_users['transparency_metric'] = (
    msft_requests_users[2015]/msft_requests_users['2015_y'] +
    msft_requests_users[2016]/msft_requests_users['2016_y'] +
    msft_requests_users[2017]/msft_requests_users['2017_y'])*1000000
```

```python
msft_requests_sorted = msft_requests_users.sort_values(by=['transparency_metric'], ascending=False)
msft_requests_sorted
```

```python
ax = msft_requests_sorted.plot.line(x='Country.1', y='transparency_metric', rot=1)
```

### In an ideal world, we would expect this graph to be normally distributed instead of a long tail

```python
msft_requests_sorted_by_users = msft_requests_users.sort_values(by=['2017_y'], ascending=False)
ax = msft_requests_sorted_by_users[:20].plot.bar(x='Country.1', y='transparency_metric')
```
